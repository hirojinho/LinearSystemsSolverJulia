```python
from datetime import datetime, timedelta
from tabulate import tabulate

# Start date
start_date = datetime(2025, 4, 20)
weeks = 12
split_point = 6  # Split into two tables of 6 weeks

# Generate week ranges
week_ranges = [
    f"{(start_date + timedelta(weeks=i)).strftime('%d/%m')} - "
    f"{(start_date + timedelta(weeks=i, days=4)).strftime('%d/%m')}"
    for i in range(weeks)
]

# Define activity hours
activities = {
    "1": [20, 20, 12, 12, 8, 8, 0, 0, 0, 0, 0, 0],
    "2": [0, 0, 8, 8, 12, 12, 12, 12, 0, 0, 0, 0],
    "3": [0, 0, 0, 0, 0, 0, 8, 8, 20, 20, 20, 20],
}

# Function to build and print one table
def print_table(start, end, title=""):
    sub_ranges = week_ranges[start:end]
    table = [["Atividade / Semana"] + sub_ranges]
    for key, hours in activities.items():
        table.append([key] + hours[start:end])
    
    # Total for this chunk
    total_hours = [sum(week) for week in zip(*[v[start:end] for v in activities.values()])]
    table.append(["Total Semana (h)"] + total_hours)

    total_activity_hours = sum(total_hours)
    table.append(["Total (h)", ""] + [total_activity_hours] + [""] * (len(sub_ranges) - 1))

    print(f"\n{title}")
    print(tabulate(table, headers="firstrow", tablefmt="grid"))
    print("\n")

# Print both tables
print_table(0, split_point, "Tabela 1: Semanas 1 a 6")
print_table(split_point, weeks, "Tabela 2: Semanas 7 a 12")

```
